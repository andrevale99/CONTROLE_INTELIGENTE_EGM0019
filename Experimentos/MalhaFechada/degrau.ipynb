```python
import matplotlib.pyplot as plt
import numpy as np 
import os
import pandas as pd

PATH = os.getcwd()
print(PATH)

print(os.listdir())

TextFile = f"./degrau/teste_22_T100ms.txt"
print(TextFile)

data = np.loadtxt(TextFile)

tempo = data[:, 0]
rpm = data[:,1]
setpoint = data[:,2]
erro = data[:,3]
acaoP = data[:,4]
acaoI = data[:,5]
acaoD = data[:,6]
sinal = data[:,7]
```

## Diposição das colunas

**1.** Tempo (s)

**2.** RPM

**3.** Setpoint

**4.** erro

**5.** ação P

**6.** ação I

**7.** ação D

**8.** Sinal

```python
def plotData(xdata, ylist, labellist=[],title="",xlabel="", ylabel="", saveImg = False, titleImg="Grafico"):
    for i in range(len(ylist)):
        plt.plot(xdata, ylist[i], label=labellist[i])

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid()

    if saveImg == True:
        plt.savefig(PATH + "/" + titleImg + ".png")

    plt.show()
    
```

```python
plotData(tempo, [setpoint],["setpoint"],
         "Tempo x Setpoint",
         "tempo (s)", "setpoint",
         False, "TempoxSetpoint")
```

```python
plotData(tempo, [setpoint, rpm],["setpoint", "rpm"],
         "Tempo x Setpoint",
         "tempo (s)", "setpoint",
         False, "TempoxSetpoint")
```

```python
plotData(tempo, [erro],["erro"],
         "Tempo x Setpoint",
         "tempo (s)", "setpoint",
         False, "TempoxSetpoint")
```

```python
plotData(tempo, [acaoD,acaoP,acaoI],["acaoD","acaoP","acaoI"],
         "Tempo x Setpoint",
         "tempo (s)", "setpoint",
         False, "TempoxSetpoint")
```
