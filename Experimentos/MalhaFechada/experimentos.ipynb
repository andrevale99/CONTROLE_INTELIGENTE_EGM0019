```python
import matplotlib.pyplot as plt
import numpy as np 
import os
import pandas as pd

PATH = os.getcwd()
print(PATH)

print(os.listdir())

TextFile = f"teste_07_T100ms.txt"
print(TextFile)

data = np.loadtxt(TextFile)

tempo = data[:, 0]
tempo_de_calculo = np.mean(data[:,1])
setpoint = data[:,2]
pwm = data[:,3]
qtdPulsosCanalA = data[:,4]
qtdPulsosPorPeriodo = data[:,5]
theta = data[:,6]
omega = data[:,7]
erro = data[:,8]
acaoP = data[:,9]
acaoI = data[:,10]
acaoD = data[:,11]
sinalDeControle = data[:,12]
```

# Disposição das colunas

**1.** Tempo (s)

**2.** Tempo Total do calculo (us)

**3.** Setpoint

**4.** PWM

**5.** Quantidade de Pulsos do canal A

**6.** Quantidade de Pulsos por Período

**7.** Theta (graus)

**8.** Omega (RPM)

**9.** Erro

**10.** Ação Proporcional

**11.** Ação Integrativa

**12.** Ação Derivativa

**13.** Sinal de Controle

```python
def plotData(xdata, ylist, title="",xlabel="", ylabel="", saveImg = False, titleImg="Grafico"):
    for i in ylist:
        plt.plot(xdata, i)

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid()

    if saveImg == True:
        plt.savefig(PATH + "/" + titleImg + ".png")

    plt.show()
    
```

# Tempo x Setpoint

```python
plotData(tempo, [setpoint],
         "Tempo x Setpoint",
         "tempo (s)", "setpoint",
         False, "TempoxSetpoint")
```

# Tempo x Omega

```python
plotData(tempo, [omega],
         "Tempo x Omega",
         "tempo (s)", "Omega (RPM)",
         False, "TempoxOmega")
```

```python
plotData(tempo, [omega, setpoint],
         "Tempo x Omega e Setpoint",
         "tempo (s)", "Omega (RPM)",
         False, "TempoxSetpoint")
```
