$Resolucao = \frac{Encoder}{\theta} \cdot 360$

```python
# linha 19222 do arquivo teste_02.txt
theta_o = 7611.49

# linha 19223 do arquivo teste_02.txt
theta_f = 7505.98

theta = (np.abs(theta_f-theta_o))

print(theta)


Resolucao =  212153130 / theta * 360
```

# Gráficos dos dados

```python
data = np.loadtxt(TextFile)

# plt.plot(data[:,0],data[:,1])
plt.plot(data[0:6500,0],data[0:6500,1])

plt.title("Tempo do Cálculo do encoder (Periodo = 100ms)")
plt.ylabel("Tempo do cálculo (us)")
plt.xlabel('s')
plt.grid()

plt.savefig("./img/tempoDeCalculo.png")

plt.show()
```

```python
print(f'Valor maximo em RPM: {np.max(data[:,8])}')
print(f'Valor maximo em RPM (absoluto): {np.abs(np.min(data[:,8]))}')

plt.plot(data[0:6500,0], data[0:6500,8], label="RPM")
plt.plot(data[0:6500,0], -1*data[0:6500,3], label="PWM")

plt.title("Omega (RPM)")
plt.legend()
plt.ylabel("RPM")
plt.xlabel('s')
plt.grid()

plt.savefig("./img/Testede")

plt.show()
```

```python
idx_inferior = 8000
idx_superior = 10100

tempo_vetor = np.copy(data[idx_inferior:idx_superior,0])
rpm_vetor = np.copy(data[idx_inferior:idx_superior,8])

media =  np.mean(rpm_vetor)
print(f'Media dos valores de RPM: {media}')

plt.plot(tempo_vetor, rpm_vetor)

plt.title("Omega (RPM) estabilização")
plt.ylabel("RPM")
plt.xlabel('s')
plt.grid()

plt.savefig("./img/RPM_estabilizacao.png")

plt.show()
```
